# 5. Scrape the details of trending repositories on Github.com.

Url = https://github.com/

You have to find the following details:

A) Repository title

B) Repository description

C) Contributors count

D) Language used


```python
url = "https://github.com/"
driver = webdriver.Chrome(r"C:/Users/siddh/Downloads/chromedriver_win32/chromedriver.exe")
driver.get(url)
driver.maximize_window()
```

```python
exp_btn = driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/summary")
try:
    exp_btn.click()
except:
    pass
```

```python
trnd_url = driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[5]/a")
urls = trnd_url.get_attribute("href")
driver.get(urls)
```

```python
# Creating Empty list

repo_urls = []
rep_title = []
description = []
contributors = []
language = []
lang = []
```

```python
# Fetching URLs for each repository

repo = driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']//a")
for i in repo:
    repo_urls.append(i.get_attribute("href"))
    
#Scrapping data of Repository Title

title = driver.find_elements_by_xpath("//h1[@class = 'h3 lh-condensed']")
for i in title:
    rep_title.append(i.text)
    
#Scrapping data from every Repository page:
for i in repo_urls:
    driver.get(i)
    time.sleep(3)

    #Scrapping data of description:
    try:
        desc = driver.find_element_by_xpath("//p[@class = 'f4 my-3']")
        description.append(desc.text)
    except:
        description.append('-')
        
    #SCrapping data of Contributors:
    try:
        cont = driver.find_element_by_xpath("//a[contains(@href,'contributors')]/span")
        contributors.append(cont.text)
    except:
        contributors.append('-')
    
    #Scrapping language used:
    try:
        for i in driver.find_element_by_xpath("//a[contains(@href,'search?')]/span[1]"):
            lang.append(i.text)
        language.append(lang)
    except:
        language.append('-')
```

```python
gith = {"Repository Title":rep_title,"Repository Description":description,"Contributors Counts":contributors,"Language Used":language}
GitHubDF = pd.DataFrame(gith)
GitHubDF.reset_index(drop = True, inplace = True)
GitHubDF.index+=1
```
