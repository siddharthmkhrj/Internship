```python
close_button = driver.find_element_by_xpath("//button[@class = '_2KpZ6l _2doB4z']")
close_button.click()

search_product = driver.find_element_by_xpath("//input [@class = '_3704LK']")
search_product.send_keys('sunglasses')

search_button = driver.find_element_by_xpath("//button [@class = 'L0Z3Pu']")
search_button.click()
```

```python
brand_name = []
product_des = []
product_price = []
```

```python
brand_tag = driver.find_elements_by_xpath("//div [@class = '_2WkVRV']")
product_des_tag = driver.find_elements_by_xpath("//a [@class = 'IRpwTa']")
price_tag = driver.find_elements_by_xpath("//div [@class = '_30jeq3']")

for i in brand_tag:
    brand = i.text
    brand_name.append(brand)

for b in product_des_tag:
    product = b.text
    product_des.append(product)
    
for c in price_tag:
    price = c.text
    product_price.append(price)
    
```

```python
import time
```

```python
i=0
nextbtn = driver.find_element_by_class_name('_1LKTO3')
while(i<2):
    try:
        nextbtn.click()
    except:
        pass
    time.sleep(3) #to allow the page to load completely before scraping.
    brand_tag = driver.find_elements_by_xpath("//div [@class = '_2WkVRV']")
    product_des_tag = driver.find_elements_by_xpath("//a [@class = 'IRpwTa']")
    price_tag = driver.find_elements_by_xpath("//div [@class = '_30jeq3']")
    
    for b in brand_tag:
        try:
            brand_name.append(b.text)
        except:
            pass
    
    for d in product_des_tag:
        try:
            product_des.append(d.text)
        except:
            pass
    
    for p in price_tag:
        try:
            product_price.append(p.text)
        except:
            pass
        
    if 'inactive' in nextbtn.get_attribute('class'):
        break;
    
    i=i+1
```

```python
sunglasses = pd.DataFrame({"Brand Name": brand_name[0:100],"Product Description":product_des[0:100], "Price":product_price[0:100]})
sunglasses
```
